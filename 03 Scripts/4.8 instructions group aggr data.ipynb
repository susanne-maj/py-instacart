```python
# Import libraries

import pandas as pd
import numpy as np
import os
```

```python
# define path
path = r'C:\Users\susan\Documents\data analytics\Instacart Basket Analysis\02 Data\Prepared data'
```

```python
# import ords_prods_merge
ords_prods_merge = pd.read_pickle(os.path.join(path, 'ords_prods_merge.pkl'))
```

```python
# Create a subset with one million records
df = ords_prods_merge[:1000000]
```

```python
# Check the dimensions
df.shape
```

```python
# Check the first columns
df.head(10)
```

```python
# Group by column product name
df.groupby('product_name')
```

```python
# Split data into groups based on dep id and obtain mean values for the order nr column
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Group by dep nr and take the mean of the order nr without the agg function
df.groupby('department_id')['order_number'].mean()
```

```python
# Use the dots instead of square brackets for grouping
df.groupby('department_id').order_number.mean()
```

```python
# Group by dep id, aggregate by order nr and determine mean, min and max
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
# Create a column for max order, group by user id, transform to show max orders per user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Show file 
ords_prods_merge.head(5)
```

```python
# Create a loyalty flag for the customers based on max order value
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Count the values with the new flag. (Only new customers????)
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Display only a few columns
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(10)
```
